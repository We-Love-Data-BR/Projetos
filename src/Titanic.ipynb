```python
# Funcao para remover as colunas
def remove(colunas):
    for i in range(0, len(colunas)):
        ds.drop([colunas[i]], axis = 1, inplace = True)
    return ds.head()
```

```python
# Relação de colunas que vamos remover
colunas = ['Ticket','Cabin','PassengerId']

# Chama a funcao e passar as colunas que serão removidas
remove(colunas)
```

```python
# Alterando o nome das colunas
ds.columns = ['Sobreviveu','Classe','Nome','Sexo','Idade','Irm_Conj','Pais_Filhos','Tarifa','Embarque']
ds.head()
```

```python
ds['Embarque'] = ds['Embarque'].replace(['S','C','Q'], ['Southampton','Queenstown','Cherbourg'])
ds['Sobreviveu'] = ds['Sobreviveu'].replace([0,1], ['Não', 'Sim'])
ds['Sexo'] = ds['Sexo'].replace(['male','female'], ['Masculino','Femino'])

ds.head(6)
```

```python
# Quantidade de sobreviventes separados por sexo
sobreviventes = sns.countplot(data=ds, x='Sobreviveu', hue='Sexo', palette='Blues_d')

# Exibindo valores no topo das barras
for p in sobreviventes.patches:
    b = p.get_bbox()
    sobreviventes.annotate("{:.0f}".format(b.y1 + b.y0), ((b.x0 + b.x1)/2 - 0.03, b.y1 + 15))

plt.title('Quantidade de Sobreviventes / Mortos\n Total: %s' % (ds.shape[0]))
plt.xlabel('Sobreviveu')
plt.ylabel('Número de Sobreviventes')
plt.show()
```
